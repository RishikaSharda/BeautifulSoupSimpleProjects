```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://pages.stern.nyu.edu/~adamodar/New_Home_Page/datafile/countrytaxrates.html"
```

```python
#False to ignore SSL certificate verification
response = requests.get(url, verify=False)
print(response.content)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
# Find the table element in the HTML content
table = soup.find("table", {"class": "maintable"})
```

```python
rows = table.find_all('tr')
```

```python
data = []
for row in rows:
    cells = row.find_all('td')
    row_data = [cell.get_text(strip=True) for cell in cells]
    data.append(row_data)
```

```python
df = pd.DataFrame(data[1:], columns=data[0])
```
